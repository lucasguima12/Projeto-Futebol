```python
import pandas as pd
import http.client
import json
from datetime import datetime
```

```python
def team_id(liga):
        conn = http.client.HTTPSConnection("api-football-v1.p.rapidapi.com")

        headers = {
            'X-RapidAPI-Key': "1fbf96808emshacec9ff25496937p10c070jsn2d5baedc8b83",
            'X-RapidAPI-Host': "api-football-v1.p.rapidapi.com"
        }

        conn.request("GET", f"/v3/teams?league={liga}&season=2023", headers=headers)

        res = conn.getresponse()
        data = res.read()
        # Decode the byte string to a UTF-8 string
        data_str = data.decode('utf-8')

        # Parse the JSON data into a Python dictionary
        data_dict = json.loads(data_str)

        # Assuming the 'response' key contains a list of fixtures
        df_times = pd.json_normalize(data_dict['response'])
        
        return df_times
    
    
```

## Funcao para visualizar as estatisticas de cada time ate o final do campeonato

```python
dataframes_list = []

def estatisticas_final():
    ligas = [39, 71, 78, 61, 135, 307, 253, 128, 140]
    for liga in ligas:
        df_times = team_id(liga)       
        for id_time in df_times['team.id']:
            conn = http.client.HTTPSConnection("api-football-v1.p.rapidapi.com")
            headers = {
                'X-RapidAPI-Key': "1fbf96808emshacec9ff25496937p10c070jsn2d5baedc8b83",
                'X-RapidAPI-Host': "api-football-v1.p.rapidapi.com"
            }

            conn.request("GET", f"/v3/teams/statistics?league={liga}&season=2023&team={id_time}", headers=headers)
            res = conn.getresponse()
            data = res.read()
            estatisicas_time = json.loads(data.decode("utf-8"))
            # Check if the 'response' key is in the dictionary
            if 'response' in estatisicas_time:
                # Normalize the 'response' data and append the DataFrame to the list
                df_estatisticas = pd.json_normalize(estatisicas_time['response'])
                dataframes_list.append(df_estatisticas)
    # Concatenate all the DataFrames into one
    df_jogadores = pd.concat(dataframes_list, ignore_index=True)
    
    return df_jogadores
```

```python
df_bruto_final = estatisticas_final()
```

```python
df_todas_ligas = df_bruto_final.dropna(axis=1, how='all')
df_todas_ligas = df_todas_ligas[['form', 'league.id', 'league.name',
        'league.season', 'team.id', 'team.name', 'fixtures.played.total','fixtures.wins.total', 'fixtures.draws.total',
                    'fixtures.loses.total', 'goals.for.total.total','biggest.streak.wins', 'biggest.streak.loses']]
```

```python
df_todas_ligas['Pontos'] = df_todas_ligas['fixtures.wins.total'] * 3 + df_todas_ligas['fixtures.draws.total'] * 1
```

```python
df_liga_inglesa = df_todas_ligas[df_todas_ligas['league.id'] == 39].sort_values(by = 'Pontos', ascending = False)
df_brasileirao = df_todas_ligas[df_todas_ligas['league.id'] == 71].sort_values(by = 'Pontos', ascending = False)
df_liga_espanhola = df_todas_ligas[df_todas_ligas['league.id'] == 140].sort_values(by = 'Pontos', ascending = False)
df_liga_italiana = df_todas_ligas[df_todas_ligas['league.id'] == 135].sort_values(by = 'Pontos', ascending = False)
df_liga_alema = df_todas_ligas[df_todas_ligas['league.id'] == 78].sort_values(by = 'Pontos', ascending = False)
df_liga_francesa = df_todas_ligas[df_todas_ligas['league.id'] == 61].sort_values(by = 'Pontos', ascending = False)
df_liga_arabe = df_todas_ligas[df_todas_ligas['league.id'] == 307].sort_values(by = 'Pontos', ascending = False)
df_liga_argentina = df_todas_ligas[df_todas_ligas['league.id'] == 128].sort_values(by = 'Pontos', ascending = False)
df_liga_americana = df_todas_ligas[df_todas_ligas['league.id'] == 253].sort_values(by = 'Pontos', ascending = False)
```

```python
df_liga_inglesa.to_csv('df_liga_inglesa.csv')
df_brasileirao.to_csv('df_brasileirao.csv')
df_liga_espanhola.to_csv('df_liga_espanhola.csv')
df_liga_italiana.to_csv('df_liga_italiana.csv')
df_liga_alema.to_csv('df_liga_alema.csv')
df_liga_francesa.to_csv('df_liga_francesa.csv')
df_liga_arabe.to_csv('df_liga_arabe.csv')
df_liga_argentina.to_csv('df_liga_argentina.csv')
df_liga_americana.to_csv('df_liga_americana.csv')
```
