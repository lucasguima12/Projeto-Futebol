```python
import pandas as pd
import http.client
import json
from datetime import datetime
```

```python
def team_id(ano):
        conn = http.client.HTTPSConnection("api-football-v1.p.rapidapi.com")

        headers = {
            'X-RapidAPI-Key': "1fbf96808emshacec9ff25496937p10c070jsn2d5baedc8b83",
            'X-RapidAPI-Host': "api-football-v1.p.rapidapi.com"
        }

        conn.request("GET", f"/v3/teams?league=71&season={ano}", headers=headers)

        res = conn.getresponse()
        data = res.read()
        # Decode the byte string to a UTF-8 string
        data_str = data.decode('utf-8')

        # Parse the JSON data into a Python dictionary
        data_dict = json.loads(data_str)

        # Assuming the 'response' key contains a list of fixtures
        df_times = pd.json_normalize(data_dict['response'])
        
        return df_times
    
    
```

## Funcao para visualizar as estatisticas de cada time ate o final do campeonato

```python
dataframes_list = []

def estatisticas_final():
    for ano in range(2010, 2024):
        df_times = team_id(ano)       
        for id_time in df_times['team.id']:
            conn = http.client.HTTPSConnection("api-football-v1.p.rapidapi.com")
            headers = {
                'X-RapidAPI-Key': "1fbf96808emshacec9ff25496937p10c070jsn2d5baedc8b83",
                'X-RapidAPI-Host': "api-football-v1.p.rapidapi.com"
            }

            conn.request("GET", f"/v3/teams/statistics?league=71&season={ano}&team={id_time}", headers=headers)
            res = conn.getresponse()
            data = res.read()
            estatisicas_time = json.loads(data.decode("utf-8"))
            # Check if the 'response' key is in the dictionary
            if 'response' in estatisicas_time:
                # Normalize the 'response' data and append the DataFrame to the list
                df_estatisticas = pd.json_normalize(estatisicas_time['response'])
                dataframes_list.append(df_estatisticas)
        # Concatenate all the DataFrames into one
    df_jogadores = pd.concat(dataframes_list, ignore_index=True)
    
    return df_jogadores
```

```python
df_bruto_final = estatisticas_final()
df_estatisticas_final = df_bruto_final.dropna(axis=1, how='all')
df_estatisticas_final = df_estatisticas_final[['form', 'league.id', 'league.name',
        'league.season', 'team.id', 'team.name', 'fixtures.played.total','fixtures.wins.total', 'fixtures.draws.total',
                    'fixtures.loses.total', 'goals.for.total.total','biggest.streak.wins', 'biggest.streak.loses']]
```

```python
df_estatisticas_final['Pontos'] = df_estatisticas_final['fixtures.wins.total'] * 3 + df_estatisticas_final['fixtures.draws.total'] * 1
```

```python
df_estatisticas_final = df_estatisticas_final.sort_values(by = 'Pontos', ascending = False)
```

## Funçao para visualizar Funcao para visualizar as estatiscas de cada time ate a 19 rodada.

### Antes precisamos obter a data do ultimo jogo da 19 rodada de cada ano

```python
def obter_data_19_rodada(ano):
    conn = http.client.HTTPSConnection("api-football-v1.p.rapidapi.com")
    headers = {
        'X-RapidAPI-Key': "1fbf96808emshacec9ff25496937p10c070jsn2d5baedc8b83",
        'X-RapidAPI-Host': "api-football-v1.p.rapidapi.com"
    }
    conn.request("GET", f"/v3/fixtures?league=71&season={ano}", headers=headers)
    res = conn.getresponse()
    data = res.read()

    # Decode the byte string to a UTF-8 string
    data_str = data.decode('utf-8')

    # Parse the JSON data into a Python dictionary
    data_dict = json.loads(data_str)

    # Assuming the 'response' key contains a list of fixtures
    df = pd.json_normalize(data_dict['response'])

    # Filtrar pela 19ª rodada
    rodada_19 = df[df['league.round'] == 'Regular Season - 19']

    # Encontrar a data do último jogo
    data_final = rodada_19['fixture.date'].max()
    
    data_obj = datetime.fromisoformat(data_final)

    # Formatando a data no formato 'yyyy-mm-dd'
    data_formatada = data_obj.strftime('%Y-%m-%d')
    
    return data_formatada
```

### Com esta funçao, podemos agora fazer a funçao que chama as estatisticas

```python
dataframes_list = []

def estatisticas_rodada_19():
    for ano in range(2010, 2024):
        data_final_rodada_19 = obter_data_19_rodada(ano)
        
        df_times = team_id(ano)
        
        for id_time in df_times['team.id']:
            conn = http.client.HTTPSConnection("api-football-v1.p.rapidapi.com")
            headers = {
                'X-RapidAPI-Key': "1fbf96808emshacec9ff25496937p10c070jsn2d5baedc8b83",
                'X-RapidAPI-Host': "api-football-v1.p.rapidapi.com"
            }

            conn.request("GET", f"/v3/teams/statistics?league=71&season={ano}&team={id_time}&date={data_final_rodada_19}", headers=headers)
            res = conn.getresponse()
            data = res.read()
            estatisicas_time = json.loads(data.decode("utf-8"))
            # Check if the 'response' key is in the dictionary
            if 'response' in estatisicas_time:
                # Normalize the 'response' data and append the DataFrame to the list
                df_estatisticas = pd.json_normalize(estatisicas_time['response'])
                dataframes_list.append(df_estatisticas)
        # Concatenate all the DataFrames into one
    df_bruto = pd.concat(dataframes_list, ignore_index=True)
    
    return df_bruto
```

```python
df_bruto_rodada_19 = estatisticas_rodada_19()
df_estatisticas_rodada_19 = df_bruto_rodada_19.dropna(axis=1, how='all')
df_estatisticas_rodada_19 = df_estatisticas_rodada_19[['form', 'league.id', 'league.name',
        'league.season', 'team.id', 'team.name', 'fixtures.played.total','fixtures.wins.total', 'fixtures.draws.total',
                    'fixtures.loses.total', 'goals.for.total.total','biggest.streak.wins', 'biggest.streak.loses']]
```

```python
df_estatisticas_rodada_19 = df_estatisticas_rodada_19.sort_values(by = 'Pontos', ascending = False)
```
